### Encoding

## Modeling

```python
x = df.drop('Status', axis=1)
y = df['Status']

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42,stratify=y)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000, class_weight="balanced")
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
import joblib
joblib.dump(model, "model_logistic.pkl")
```

## Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

Model menghasilkan akurasi sebesar 73% dalam memprediksi status mahasiswa (dropout, enrolled, atau graduate) hal ini ditunjukkan oleh nilai precision dan recall yang relatif tinggi pada kedua kelas tersebut. Namun, performa pada kelas enrolled masih lebih rendah dibandingkan kelas lainnya, terutama pada nilai precision. Hal ini menunjukkan bahwa model masih menghasilkan cukup banyak prediksi enrolled yang tidak sesuai dengan kondisi sebenarnya.
